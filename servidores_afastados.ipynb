# Afastamento de servidores federais desde o começo da pandemia da Covid-19 no país

Base de dados: https://dados.gov.br/dataset/afastamento-remunerado

```python
#importando bibliotecas
import pandas as pd
import glob
```

```python
#caminho dos arquivos
files = glob.glob(r'C:\Users\judite.cypreste\servidores_afastados\*')
```

```python
#me mostre tudo
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
#pegando o primeiro csv como modelo
df = pd.read_csv(r'.\servidores_afastados\AFASTREM_012020.csv', encoding='Latin5',sep=';')
```

```python
#unindo os csvs
df = pd.concat([pd.read_csv(file, encoding='Latin5', sep=';', error_bad_lines=False, warn_bad_lines=True) for file in files])
```

```python
#vendo meu df
df.head()
```

```python
#analisando quais são as descrições para afastamento disponíveis
df['Descrição do afastamento'].unique()
```

```python
#substituindo ',' por '.' e dizendo que a coluna 'Valor Transação' é numerica
df['Valor rendimento liquido'] = df['Valor rendimento liquido'].str.replace(',','.')
df['Valor rendimento liquido'] = df['Valor rendimento liquido'].astype(float)
```

## Analisando quem ficou em Home Office

```python
#separando aqueles que estão conseguindo realizar suas atividade de maneira 'home office'
home_office = df[df['Descrição do afastamento'].isin(['Trabalho Remoto COVID-19                                    ', 'TRABALHO REMOTO COVID-19                                    '])]
```

```python
#calculando quanto foi gasto para servidores em home office
gastos_ho = home_office['Valor rendimento liquido'].sum()
gastos_ho
```

```python
#se um servidor ficou mais de um mês em casa, ele aparecerá mais de uma vez na base. por isso, vamos retirar os duplicados para sabermos quantos funcionários ficaram em home office
home_office.drop_duplicates(subset = ("Nome", 'CPF'), keep = False, inplace = True)
```

```python
#contando quantos servidores estão ficaram em home office
home_office.count()
```

## Analisando quem ficou sem poder trabalhar e recebeu mesmo assim

```python
#separando aqueles que não conseguiram realizar nenhuma atividade em casa e ficaram apenas recebendo salários
afastados = df[df['Descrição do afastamento'].isin(['AFASTAMENTO COVID-19                                        ', 'Afastamento COVID-19                                        '])]
```

```python
#calculando quanto foi gasto para servidores afastados
gastos_afastados = afastados['Valor rendimento liquido'].sum()
gastos_afastados
```

```python
#se um servidor ficou mais de um mês em casa, ele aparecerá mais de uma vez na base. por isso, vamos retirar os duplicados para sabermos quantos funcionários ficaram afastados
afastados.drop_duplicates(subset = ("Nome", 'CPF'), keep = False, inplace = True)
```

```python
#contando quantos servidores estão ficaram afastados
afastados.count()
```

```python
#qual foi o cargo que mais teve afastamentos contabilizados 
por_cargo = afastados.groupby('Descrição do cargo emprego')['Descrição do cargo emprego'].count().sort_values(ascending=False)
por_cargo
```

```python
#qual foi o órgão que mais afastou funcionários
por_orgao = afastados.groupby('Nome do órgão de origem')['Nome do órgão de origem'].count().sort_values(ascending=False)
por_orgao
```
